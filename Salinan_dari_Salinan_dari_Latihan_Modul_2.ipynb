## 3. Select, From, dan Where Statement

```python
# menampilkan semua data dari tabel orderdetails
query_where = """

SELECT * FROM OrderDetails

"""
pd.read_sql_query(query_where, conn)
```

```python
# menampilkan semua data dari tabel orderdetails yang ada diskonnya
query_where = """

SELECT * FROM orderdetails
WHERE Discount > 0.00

"""
pd.read_sql_query(query_where, conn)
```

```python
# mengambil 10 data dari tabel orderdetails
query_limit = """

SELECT * FROM orderdetails
LIMIT 10

"""
pd.read_sql_query(query_limit, conn)
```

## 4. Operator

```python
# menghitung harga asli setelah dikurangi diskon
query_order = """

SELECT OrderID, ProductID,
       UnitPrice * (1-Discount) AS RealPrice,
       Quantity
FROM orderdetails
LIMIT 10

"""
pd.read_sql_query(query_order, conn)
```

```python
# menampilkan data yang memiliki quantity > 30
query_where = """

SELECT productid as product_id, quantity FROM orderdetails
WHERE quantity > 30
"""
pd.read_sql_query(query_where, conn)
```

```python
# menampilkan data yang memiliki harga < 10 dan diskon > 2%
query_where = """

SELECT * FROM orderdetails
WHERE UnitPrice < 10 AND Discount > 0.20
"""
pd.read_sql_query(query_where, conn)
```

## 5. Basic Function


```python
# menampilkan contoh basic function
query_order = """

SELECT ProductID,
       SUM(UnitPrice),
       SUM(Quantity),
       COUNT(Quantity),
       COUNT(OrderID)
FROM orderdetails
GROUP BY 1

"""
pd.read_sql_query(query_order, conn)
```

```python
# menampilkan maximum harga product / produk termahal
query_order = """

SELECT ProductID,
       MAX(UnitPrice)
FROM orderdetails

"""
pd.read_sql_query(query_order, conn)
```

```python
# menampilkan rata-rata harga per produk
query_order = """

SELECT ProductID, AVG(UnitPrice) as RataRata_Harga
FROM orderdetails
GROUP BY 1

"""
pd.read_sql_query(query_order, conn)
```

```python
# menampilkan rata-rata harga produk
query_order = """

SELECT AVG(UnitPrice) as RataRata_Harga
FROM orderdetails

"""
pd.read_sql_query(query_order, conn)
```

# 6. Study Case

```python
# tampilkan data dari tabel customers dengan tujuan negara Londen dengan title sebagai Sales Representative
query_cust = """
SELECT * FROM Customers
WHERE Country = 'Londen' OR ContactTitle = 'Sales Representative';
"""
pd.read_sql_query(query_cust, conn)
```

```python
# tampilkan data dari tabel categories yang mengandung coffees di dalamnya
query_category = """
SELECT * FROM Categories
WHERE Description LIKE '%coffees%';
"""
pd.read_sql_query(query_category, conn)
```

```python
# tampilkan data dari tabel products yang masih ada (belum discontinued) dengan rentang harga antara 10 sampai 15
query_product = """
SELECT * FROM Products
WHERE Discontinued = 0 AND UnitPrice BETWEEN 10 AND 15;
"""
pd.read_sql_query(query_product, conn)
```

```python
# hitung jumlah data customer di yang tinggal di Jerman
query_count = """
SELECT COUNT(*) AS CustomerCount FROM Customers
WHERE Country = 'Germany';
"""
pd.read_sql_query(query_count, conn)
```

```python
# berapakah harga termurah dengan productid  46 jika semuanya mendapatkan diskon sebesar 27% pada tabel orderdetails
query_min_price = """
SELECT MIN(UnitPrice * (1 - Discount)) AS MinPrice
FROM OrderDetails
WHERE ProductID = 46;
"""

# Menjalankan query dan membaca hasilnya ke dalam DataFrame
df_min_price = pd.read_sql_query(query_min_price, conn)

# Menampilkan harga termurah setelah diskon
print("Harga termurah setelah diskon untuk produk dengan ProductID 46:", df_min_price['MinPrice'][0])
```
